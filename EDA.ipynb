```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6) 
```

```python

df = pd.read_csv('predictive_criminal_justice_dataset.csv')

print(df.head())
print("Information about the dataset:")
print(df.info())
```

```python
print(df.describe())
duplicate_count = df.duplicated().sum()
print(f"Number of duplicates are: {duplicate_count}")
```

```python
plt.figure(figsize = (12,6))
sns.countplot(data = df, x = "Case_Outcome")
plt.title("Distribution of Case Outcomes")
plt.xlabel("Case Outcome")
plt.ylabel("Count")
plt.show()
```

```python
fig, axes = plt.subplots(1,2, figsize=(15,6))
sns.boxplot(ax = axes[0], data = df, x = "Case_Outcome", y = "Prior_Convictions")
axes[0].set_title("Prior Convictions by Case Outcome")

sns.boxplot(ax = axes[1], data = df, x = "Case_Outcome", y = "Risk_Score")
axes[1].set_title("Risk Score by Case Outcome")
```

```python
plt.figure(figsize = (12, 6))
sns.countplot(data = df, x = "Crime_Type", hue = "Case_Outcome")
plt.title("Distribution of Case Outcomes by Crime Type")
plt.xlabel("Crime Type")
plt.ylabel("Count")
plt.xticks(rotation=0) # Rotate x-axis labels for better readability
plt.show()
```
